```python
from urllib.request import urlopen

import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
from albumentations import HorizontalFlip, IAAPerspective, ShiftScaleRotate
from albumentations.example_transform import strong_aug
```

```python
def download_image(url):
    data = urlopen(url).read()
    data = np.frombuffer(data, np.uint8)
    image = cv2.imdecode(data, cv2.IMREAD_COLOR)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    return image
```

```python
image = download_image('https://d177hi9zlsijyy.cloudfront.net/wp-content/uploads/sites/2/2018/05/11202041/180511105900-atlas-boston-dynamics-robot-running-super-tease.jpg')
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(image)
plt.show()
```

```python
def augment_and_show(aug, image):
    image = aug(image=image)['image']
    plt.figure(figsize=(10, 10))
    plt.imshow(image)
```

```python
aug = HorizontalFlip(p=1)
augment_and_show(aug, image)
```

```python
aug = IAAPerspective(scale=0.2, p=1)
augment_and_show(aug, image)
```

```python
aug = ShiftScaleRotate(p=1)
augment_and_show(aug, image)
```

```python
aug = strong_aug()
augment_and_show(aug, image)
```
